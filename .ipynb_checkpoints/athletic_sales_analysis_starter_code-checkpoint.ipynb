```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df2020 = pd.read_csv("Resources/athletic_sales_2020.csv")
df2021 = pd.read_csv("Resources/athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
df2020.head()
```

```python
# Display the 2021 sales DataFrame
df2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
df2020.describe()
```

```python
# Check the 2021 sales data types.
df2021.describe()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
result = pd.concat([df2020, df2021], axis=0, join='inner').reset_index(drop=True)
result.tail()
```

```python
# Check if any values are null.
null_values = result.isnull().any()
print("Null Values in Each Column:\n", null_values)
```

```python
# Check the data type of each column
data_types_before = result.dtypes
print("\nData Types Before Conversion:\n", data_types_before)
```

```python
# Convert the "invoice_date" to a datetime datatype
result['invoice_date'] = pd.to_datetime(result['invoice_date'], format='%m/%d/%Y')

```

```python
# Confirm that the "invoice_date" data type has been changed.
data_types_after = result.dtypes
print("\nData Types After Conversion:\n", data_types_after)
```
